```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
file_path = '/Users/markduan/duan/USC_course/USC_APDS/INF553/homework/hw5/data/test1/data0.txt'
```

```python
df = pd.read_table(file_path, sep=',', header=None)
```

```python
X = df.iloc[:,1:]
```

```python
estimator = KMeans(n_clusters=50, verbose=True, init='random')
```

```python
estimator.fit(X)
```

```python
label_pred = estimator.labels_
```

```python
for i in range(len(label_pred)):
    print('"%d": %d' % (i, label_pred[i]))
```

```python
d = {}
for i, x in enumerate(label_pred):
    if x not in d.keys():
        d[x] = [i]
    else:
        d[x].append(i)
```

```python
for k in d:
    print((k, len(d[k])))
```
