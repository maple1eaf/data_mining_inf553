```python
from pyspark import SparkConf, SparkContext, StorageLevel
```

```python
file_path = "file:///Users/markduan/duan/USC_course/USC_APDS/INF553/project/data/train_review.json"
```

```python
conf = SparkConf() \
    .setAppName("task") \
    .setMaster("local[*]") \
    .set("spark.driver.memory","4g")
sc = SparkContext(conf=conf)
sc.setLogLevel("ERROR")
```

```python
raw_data = sc.textFile(file_path) \
    .map(json.loads) \
    .map(lambda x: (x['business_id'], x['user_id'])) \
    .persist()
```

```python
num_bid = raw_data.map(lambda x: (x[0], 1)).reduceByKey(lambda x, y: x + y).count()
```

```python
num_uid = raw_data.map(lambda x: (x[1], 1)).reduceByKey(lambda x, y: x + y).count()
```

```python
num_uid
```

```python
a = raw_data.map(lambda x: (x[0], 1)).reduceByKey(lambda x, y: x + y).map(lambda x: num_uid - x[1]).collect()
```

```python
k = sum(a)
```

```python
sparsity = k / (num_bid * num_uid)
```
